# Importing Dependencies

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

# Loading Boston Dataset

```python
from sklearn.datasets import load_boston
boston= load_boston()
print(boston)
```

# Converting into Dataframe

```python
df_x=pd.DataFrame(data=boston.data, columns=boston.feature_names)
```

```python
df_x.head()
```

```python
df_y = pd.DataFrame(boston.target)
df_y.head()
```

```python
df_x.describe()
```

# Initialize Linear Regression Model

```python
reg = linear_model.LinearRegression()
```

# Splitting into Test and Train Data

```python
x_test, x_train, y_test, y_train = train_test_split(df_x, df_y, test_size=0.33, random_state = 41)
```

# Training Data

```python
reg.fit(x_train, y_train)
```

# Print the Coefficients

```python
print(reg.coef_)
```

# Printing the Prediction for x_test

```python
y_predict = reg.predict(x_test)
```

```python
print(y_predict)
```

# Printing the real values

```python
print(y_test)
```

# Check the Model's accuracy

```python
print(np.mean((y_predict - y_test)**2))
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print(mean_squared_error(y_test, y_predict))
```

```python
plt.scatter(y_test, y_predict)
```
